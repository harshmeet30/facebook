```python
from facebook_scraper import _get_group_posts


groups=['freeCodeCamp Earth'

,'Python Developers'

,'Artificial intelligence and Machine Learning'

,'Artificial intelligence, Machine learning, Deep learning'

,'Math for Artificial Intelligence & Big Data'

,'Python Machine Learning & Deep Learning'

,'Data Science - R & Python'

,'Deep Learning and Machine Learning'

,'Python Developers'
]

id=[
'groups/BigDataPakistan'

]

print("Groups")
groupfb_data=[]
for element in id:
    for post in _get_group_posts(element): 
        post_date = (post['time'])
        temp_dict=dict(post)
        groupfb_data.append(temp_dict)
    
print("Out of group for")
print(groupfb_data)
l=groupfb_data
```

```python
import pandas as pd
df=pd.DataFrame()
```

```python
m,n,o,p,q,r,s,t,u,v,w=[],[],[],[],[],[],[],[],[],[],[]
for i in range(len(l)):
    m.append(l[i]['post_id'])
    n.append(l[i]['text'])
    o.append(l[i]['post_text'])
    p.append(l[i]['shared_text'])
    q.append(l[i]['time'])
    r.append(l[i]['image'])
    s.append(l[i]['likes'])
    t.append(l[i]['comments'])
    u.append(l[i]['shares'])
    v.append(l[i]['post_url'])
    w.append(l[i]['link'])
df['post_id']=m
df['text']=n
df['post_text']=o
df['shared_text']=p
df['time']=q
df['image']=r
df['likes']=s
df['comments']=t
df['shares']=u
df['post_url']=v
df['link']=w
    
```
